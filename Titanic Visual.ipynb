Import Libraries

We import necessary libraries for data analysis and visualization.

pandas → data handling
numpy → numerical operations
matplotlib & seaborn → visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

Load the Titanic dataset into a pandas DataFrame for analysis.

```python
df = pd.read_csv("titanic-Data.csv")
df.head()
```

Data Cleaning
🔹 3.1 Handle Missing Values

The 'Age' column contains missing values.
We replace them with the mean age to maintain data consistency.

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

Drop Irrelevant Columns

The 'Cabin' column has too many missing values and is not useful for analysis, so we remove it.

```python
df.drop(columns=['Cabin'], inplace=True)
```

Feature Engineering
🔹 4.1 Create Age Groups

We divide passengers into age groups (Child, Teen, Adult, etc.)
This helps in better analysis of survival patterns.

```python
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

🔹 4.2 Create Family Size

Family size is calculated by adding:

SibSp (siblings/spouses)
Parch (parents/children)

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

Exploratory Data Analysis (EDA)
🔹 5.1 Survival Rate by Age Group

We analyze how survival rate varies across different age groups.

```python
age_survival = df.groupby('AgeGroup')['Survived'].mean()
age_survival
```

🔹 5.2 Survival Rate by Embarkation Port

We check survival rate based on embarkation ports (C, Q, S).

```python
embark_survival = df.groupby('Embarked')['Survived'].mean()
embark_survival
```

🔹 5.3 Survival Rate by Family Size

We analyze how family size impacts survival.

```python
family_survival = df.groupby('FamilySize')['Survived'].mean()
family_survival
```

Visualizations
📊 6.1 Age Distribution

This histogram shows the distribution of passengers' ages.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

📊 6.2 Correlation Heatmap

The heatmap shows relationships between numerical features.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

📊 6.3 Survival by Family Size

This bar chart shows how survival rate changes with family size.

```python
plt.figure(figsize=(8,5))
family_survival.plot(kind='bar')
plt.title("Survival Rate by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.show()
```

Conclusion

Missing values in the dataset were handled using mean imputation.
Passengers were grouped into age categories for better analysis.
Survival rate was higher among children and younger passengers.
Passengers with small family sizes had better survival chances.
Embarkation port also influenced survival, possibly due to class differences.
Visualizations helped in understanding patterns and relationships clearly.

Overall, this analysis shows how factors like age, family size, and embarkation point affected survival during the Titanic disaster.
